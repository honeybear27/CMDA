```python
from bs4 import BeautifulSoup as bs
import requests
r = requests.get("http://blacksburg.craigslist.org/")
data = r.text
```

```python
soup = bs(data)
```

```python
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
urls = [link.get('href') for link in soup.find_all('a')]
```

```python
urls
```

```python
import pandas as pd
pa = pd.DataFrame(urls, columns = ['links'])
```

```python
pa
```

```python
l = soup.find_all(class_="act")
url_act = [link.get('href') for link in l]
url_act
```

```python
act = pd.DataFrame(url_act, columns = ['links'])
```

```python
act
```
